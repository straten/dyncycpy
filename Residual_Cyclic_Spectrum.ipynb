## Residual Cyclic Spectrum

This notebook is used to explore the residual cyclic spectra output by the `complex_cyclic_merit_lag` method when `CS.dump_residual` is set `True`

```python
from scipy.fft import rfft, fft, fftshift, ifft, fftn, ifftn
from scipy import signal
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import numpy as np
import psrchive
from pycyc import align_to_neighbour

def freq2time(hf, axis=0):
    return ifft(hf, axis=axis, norm="ortho")

def time2freq(ht, axis=0):
    return fft(ht, axis=axis, norm="ortho")
```

```python
import pickle
with open ("simulated_9/complex_cyclic_merit_lag_residual_000.pkl", "rb") as fh:
    residual = pickle.load(fh)
print (f"residual shape={residual.shape}")
```

```python
im = plt.imshow(np.real(residual), aspect="auto", origin="lower", cmap="cubehelix_r", interpolation='none')
plt.colorbar(im)
plt.show()
im = plt.imshow(np.imag(residual), aspect="auto", origin="lower", cmap="cubehelix_r", interpolation='none')
plt.colorbar(im)
plt.show()
im = plt.imshow(np.log(np.abs(residual)+1e-16), aspect="auto", origin="lower", cmap="cubehelix_r", interpolation='none')
plt.colorbar(im)
plt.show()
plt.plot(np.real(residual[0]))
plt.show()
plt.plot(np.real(residual[1]))
plt.show()
```

```python
residual=ifft(residual)
```

```python
im = plt.imshow(np.real(residual), aspect="auto", origin="lower", cmap="cubehelix_r", interpolation='none')
im = plt.imshow(np.log(np.abs(residual)+1e-16), aspect="auto", origin="lower", cmap="cubehelix_r", interpolation='none')
plt.colorbar(im)
plt.show()
for i in range(7):
  plt.plot(np.real(residual[i]))
  plt.show()
```
